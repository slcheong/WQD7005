```python
##Upload data to Hive from csv file
from pyhive import hive
import glob

#connect to hive
con = hive.connect(host='localhost',port=10000,username='hiveuser')
cur = con.cursor()


##Create a table using stock code, and upload csv file into it
def uploadHive(fileLocation,conn):
    #create table name with stock code + HP
    tableName = fileLocation[23:27] +"_HP"
    #SQL to create table
    createTableSql = 'CREATE TABLE IF NOT EXISTS KLSE.'+tableName+' (tranDate DATE, open DECIMAL(10,4), high DECIMAL(10,4), low DECIMAL(10,4), close DECIMAL(10,4), adjClose DECIMAL(10,4), volume INT) row format delimited fields terminated by ","'
    #SQL to upload csv file to relevant table
    uploadCsvSQL = "LOAD DATA LOCAL INPATH 'file:/home/student/Desktop/WQD7005/" + fileLocation + "' OVERWRITE INTO TABLE KLSE." + tableName 
    
    #Execute SQLs
    cur.execute(createTableSql)
    cur.execute(uploadCsvSQL)

    
##List of CSV file downloaded
lists = glob.glob("Milestone 1/stockPrice/*.csv")

#For each file, create a table and upload the data
for a in lists:
    uploadHive(a,cur)
```
